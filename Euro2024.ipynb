```python
team_selected=[]
def events_for_country(team_selected,fltered_data):
    team_data=filtered_data[filtered_data['team']==team_selected].reset_index(drop=True)
    return team_data
    
    
```

```python
team_selected='Spain'
team=events_for_country(team_selected,filtered_data)
```

```python
team
```

```python
df=team
def get_pass_events(df):
   pass_events=df[df['type'] == 'Pass']
   return pass_events 

pass_event=get_pass_events(df)    
    
```

```python
pass_event
```

```python
def get_progressive_passes(start,end,treshold=15):
    start_x,start_y=start
    end_x,end_y=end
    return end_x-start_x>treshold


pass_event['is_progressive']=pass_event.apply(lambda x:get_progressive_passes(x['location'],x['pass_end_location']),axis=1)

progressive_passes=pass_event[pass_event['is_progressive']]
    
```

```python
progressive_passes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch, Sbopen, VerticalPitch,FontManager





background_color = '#0d1117'
line_color = '#58a6ff'
arrow_color = '#f97316'
title_color = '#58a6ff'

pitch = Pitch(pitch_color=background_color, line_color=line_color, pitch_type='statsbomb',line_zorder=2,stripe=False)
fg, ax = pitch.draw()



for idx,row in progressive_passes.iterrows():
    start_x,start_y=row['location']
    end_x,end_y=row['pass_end_location']
    pitch.arrows(start_x,start_y,end_x,end_y,width=2, headwidth=3,headlength=3,color=arrow_color,ax=ax)
    
    
ax.set_title("Progressive Passes", fontsize=20, color=title_color, fontweight='bold')


fg.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)
ax.spines['top'].set_color(line_color)
ax.spines['bottom'].set_color(line_color)
ax.spines['left'].set_color(line_color)
ax.spines['right'].set_color(line_color)
ax.xaxis.label.set_color(line_color)
ax.yaxis.label.set_color(line_color)
ax.tick_params(colors=line_color, which='both')
plt.show()

```

PASS NETWORKS





```python
#creating a pass network using the progressive passes dataframe

pass_network=progressive_passes.groupby(['player','pass_recipient']).size().reset_index(name='pass_count')
progressive_passes[['start_x','start_y']]=pd.DataFrame(progressive_passes['location'].tolist(),index=progressive_passes.index)
player_avg_location = progressive_passes.groupby('player').agg({'start_x': 'mean', 'start_y': 'mean'}).reset_index()

```

```python
background_color = '#0d1117'
line_color = '#58a6ff'
arrow_color = '#f97316'
title_color = '#58a6ff'

# Create a pitch
pitch = Pitch(pitch_color=background_color, line_color=line_color, stripe=False)
fig, ax = pitch.draw(figsize=(10, 7))

# Plot the average player positions
for idx, row in player_avg_location.iterrows():
    ax.scatter(row['start_x'], row['start_y'], s=100, color='white', edgecolors='black', zorder=3)
    ax.text(row['start_x'], row['start_y'], row['player'], fontsize=12, ha='center', color='white', zorder=4)

# Plot the passes
for idx, row in pass_network.iterrows():
    # Get the starting and ending player positions
    start_player = player_avg_location[player_avg_location['player'] == row['player']]
    end_player = player_avg_location[player_avg_location['player'] == row['pass_recipient']]
    
    # Check if both players are present in player_avg_location
    if not start_player.empty and not end_player.empty:
        start_x = start_player['start_x'].values[0]
        start_y = start_player['start_y'].values[0]
        end_x = end_player['start_x'].values[0]
        end_y = end_player['start_y'].values[0]
        
        # Draw the arrow
        ax.arrow(start_x, start_y, end_x - start_x, end_y - start_y, color=arrow_color, width=0.5, head_width=3, head_length=3, length_includes_head=True, zorder=2)

# Customize the appearance
fig.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)
ax.set_title("Pass Network", fontsize=20, color=title_color, fontweight='bold')

plt.show()
```

```python
starting_players=team[(team['minute']==0)]['player'].dropna().unique()


substitution_minutes=team[(team['type']=='Substitution')& (team['player'].isin(starting_players))].groupby('player')['minute'].min().to_dict()


def filter_events(row):
    player=row['player']
    minute=row['minute']
    for player in substitution_minutes:
        return minute<=substitution_minutes[player]
    return True    


filtered_data=team[team['player'].isin(starting_players) & team.apply(filter_events,axis=1)]
```
